```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xi = 2
yi = -1
wc = 5
eta = 1

def L(w):
    return 1/2 * (1 - w * xi)**2

def dw_L(w):
    return -(1 - w * xi) * xi

def update(w):
    return w - eta * dw_L(w)

x = range(20)
y = []

for _ in x:
    y.append(wc)
    wn = update(wc)
    wc = wn
    
X = np.array(x)
Y = np.array(y)

plt.plot(X, Y)
plt.ylim(-1000,1000)
```
