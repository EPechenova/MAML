### Indexing and slicing of arrays

```python
a = np.arange(10)
a
```

Indices start at 0 and -1 is denotes the last one:

```python
a[0], a[2], a[-1]
```

Other that refering to particular entries of an array, there is a neat syntax to infer subarrays from given array. This is called slicing. One can specify ranges, e.g., here only the third (counting from 0) and the eighth entry:

```python
a[2:8]
```

We can also step by 2 instead of 1:

```python
a[2:8:2]
```

Start the array at the third entry:

```python
a[2:]
```

Or reverse the array:

```python
a[::-1]
```

```python
a = np.diag(np.arange(3))
a
```

```python
a[2, 1] = 10 # third line, second column
a
```

```python
a[1,:]
```

```python
a[:,1]
```

### Shaping


Create an array:

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
a
```

and flatten it:

```python
a.ravel()
```

Note that for flattening one needs a convention which of the (potentially multiple) array indices changes fastest. The default is to always use the last index to change fastest. See `Numpy` manual: [numpy.ravel](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ravel.html?highlight=ravel#numpy.ravel).

transpose it:

```python
a.T
```

flatten it again:

```python
a.T.ravel()
```

and finally reshape it:

```python
b = a.ravel()
b = b.reshape((2, 3))
b
```

Again a convention is needed which of the (potentially multiple) array indices changes fastest and the default is to always use the last one. See `Numpy` manual [numpy.reshape](https://docs.scipy.org/doc/numpy/reference/generated/numpy.reshape.html?highlight=reshape#numpy.reshape).

### Operations on Arrays

Basic operations are elementwise. Since this only works if elements have the same size `Numpy` broadcasts arrays to the same size. Regard the following example:

```python
a = np.tile(np.arange(0, 40, 10), (3, 1)).T
a
```

```python
b = np.array([0, 1, 2])
b
```

```python
a + b
```

Broadcasting can be at the same time as helpful as confusing if one is not entirely clear about the rules which are nicely explained here: [Numpy Manual](https://docs.scipy.org/doc/numpy-1.13.0/user/basics.broadcasting.html) and with nice visualizations [Scipy Lecture Notes](https://www.scipy-lectures.org/intro/numpy/operations.html#broadcasting).

## Plotting
For plotting we use the module `Matplotlib`:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

The magic `%matplotlib inline` is used to tell `Jupyter` to display the plots inline.

```python
x = np.linspace(0, 3, 20)
y = np.linspace(0, 9, 20)
plt.plot(x, y)
```

```python
plt.plot(x, y, 'o')
```

Note that other than in symbolic CAS like `Mathematica` you need to be sure to provide enough resolution in your `x` and `y` coordinate arrays in order to display a smooth functions sufficiently smoothly. Consider the next two cells:

```python
x = np.linspace(0, 2*np.pi, 5)
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
```

There are many types of plots. Have a looks at the [`matplotlib` gallary](https://matplotlib.org/gallery/index.html). Here are some of 2d plots we will employ:

```python
x = np.random.rand(10)
y = np.random.rand(10)
plt.scatter(x, y, marker='x')
```

```python
x = np.linspace(-4*np.pi, 4*np.pi, 50)
y = np.linspace(-4*np.pi, 4*np.pi, 50)

X, Y = np.meshgrid(x, y)
Z = np.sin(X*Y)
plt.contour(Z)
```

```python
plt.contourf(Z)
```
